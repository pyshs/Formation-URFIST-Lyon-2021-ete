```python
from mediawiki import MediaWiki
wikipedia = MediaWiki(user_agent='pyMediaWiki-User-Agent-String')
wikipedia.language = 'fr'
```

```python
p = wikipedia.page('CA 125')
```

```python
p = wikipedia.page('Liste de cancers')
corpus = {}
for i in p.links:
    try:
        t = wikipedia.page(i)
        corpus[i] = t.links
    except:
        print(i)
```

```python
total = []
for i in corpus:
    total+=corpus[i]
```

```python
len(total)
```

```python
len(set(total))
```

```python
import pandas as pd
pd.Series(total).value_counts()[0:10]
```
